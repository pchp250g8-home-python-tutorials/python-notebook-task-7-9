Найти целое положительное n из уравнения (3+6+9+...+3*(n-1))+(4+5.5+7+...+(8+3*n) / 2)=137.

```python
from sympy import Eq
from sympy import solve
from sympy import symbols
from math import *

n = symbols("n")
S1 = (0+3*(n - 1))*n / 2
S2 = (5.5+(8+3*n) / 2) * n / 2
eq1 = Eq(S1+S2+4, 137)
print(eq1.lhs,"=",eq1.rhs)
sol = solve(eq1, n)
for x in sol:
    x1 = float(x)
    sign = (x1 > 0) - (x1 < 0)
    fract = x1 - floor(x1)
    if(sign > 0) and (fract == 0):
        print("Positive Integer Solution:", x1)
        
```
